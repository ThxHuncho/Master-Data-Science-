#### import data

```python
data = pd.read_csv('./wdbc.csv', sep=',',header=None)
data
```

```python
X = data.iloc[:, 2:].values
labels = data.iloc[:, 0].values
```

## Normalisation 

```python
# Importation des fonctions
from sklearn.preprocessing import StandardScaler
```

```python
# Normalisation des données
SS=StandardScaler()
SS.fit(X)
Xnorm=SS.transform(X)
```

## CLUSTERING

## First Kmeans

```python
# Importation de la library
from sklearn.cluster import KMeans
k = 50
kmeans = KMeans(n_clusters=k)
```

```python
# fitting
kmeans.fit(Xnorm)
```

```python
clustering = kmeans.predict(Xnorm)
```

```python
clustering
```

```python
# centre des clusters
X_center=kmeans.cluster_centers_
```

### Final Kmeans

```python
good_init = clf.centroids_
kmeans = KMeans(n_clusters=2, init=good_init)
```

```python
kmeans.fit(Xnorm)
Xtransform=kmeans.transform(Xnorm)
Final_Label=kmeans.labels_
```

```python
import plotly.express as px
fig = px.scatter(Xtransform, 
                 x=Xtransform[:, 0], 
                 y=Xtransform[:, 1],
                 color=Final_Label,
                )
fig
```
